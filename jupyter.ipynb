```python
def T_hscaling (v):
    A = np.array([[2,1],[0,1]])
    w = A @ v
    print(f"A: \n{A}")
    return w

# arr = np.array([[2,2], [0,1]])

# print(f"mengubah nilai x tanpa mengubah nilai y: \n{T_hscaling(arr)}")
```

```python
def transform_vectors (T, v1, v2):
    matriks = np.hstack((v1, v2))
    W = T(matriks)

    print(f"matriks: \n{matriks}")
    
    return W

v1 = np.array([[1],[0]])
v2 = np.array([[0], [1]])

print()

hasil_transform =  transform_vectors(T_hscaling, v1=v1, v2=v2)

print(f"vector v1: \n{v1} \nvector2: \n{v2} \n hasil transform: \n{hasil_transform}")
```
